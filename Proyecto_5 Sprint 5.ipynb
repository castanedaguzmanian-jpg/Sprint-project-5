[Calcula los ingresos mensuales por usuario (resta el límite del paquete gratuito del número total de llamadas, mensajes de texto y datos; multiplica el resultado por el valor del plan de llamadas; añade la tarifa mensual en función del plan de llamadas). Nota: Dadas las condiciones del plan, ¡esto podría no ser tan trivial como un par de líneas! Así que no pasa nada si dedicas algo de tiempo a ello.]

````python
# Calcula el ingreso mensual para cada usuario
def calcular_ingresos_mensuales(row):
    # Pago base mensual
    ingreso_mensual = row['usd_monthly_pay']
    
    # Calcular excesos y cargos adicionales
    # Minutos extra
    minutos_extra = max(0, row['minutes_used'] - row['minutes_included'])
    ingreso_mensual += minutos_extra * row['usd_per_minute']
    
    # Mensajes extra
    mensajes_extra = max(0, row['messages_count'] - row['messages_included'])
    ingreso_mensual += mensajes_extra * row['usd_per_message']
    
    # Datos extra (convertir MB a GB)
    gb_usados = row['mb_used'] / 1024
    gb_extra = max(0, gb_usados - row['gb_per_month_included'])
    # Redondear hacia arriba los GB extra
    import math
    gb_extra_redondeados = math.ceil(gb_extra)
    ingreso_mensual += gb_extra_redondeados * row['usd_per_gb']
    
    return ingreso_mensual

# Aplicar la función a cada fila
datos_mensuales['monthly_revenue'] = datos_mensuales.apply(calcular_ingresos_mensuales, axis=1)

print("Primeros 5 registros con ingresos calculados:")
print(datos_mensuales[['user_id', 'month', 'plan', 'monthly_revenue']].head())
````
